Руководство поставило задачу оценить вклад rich-контена и медиа в продажи за прошедший месяц: необходимо предложить модель атрибуции и составить модель, которая обосновывает необходимость (или отсутствие необходимости) производства нового контента.

## Вклад медиа

Мною была бы предложена атрибуция на основе данных. В ней бы мы смогли оценить все каналы взаимодействия с пользователем и сравнить закономерности, которые привели к покупке или, наоборот, ей помешали

Также было бы необходимо установить окно конверсии. Я бы предложил считать окном 3 дня, все же принять решение о заказе продуктов на дом дается не слишком долго. (Также можно было бы рассмотреть вариант и с окном в 1 день, т.к. такое решение можно принять и спонтанно, тем более что минимальная сумма заказа невелика)

Нам бы потребовались данные о взаимодействии клиентов с различными каналами рекламы: клики, просмотры, время совершения действия, конверсии и т. д.  
Таким образом можно оценить сколько кликов по различным источникам приводят к продажам, и с каких источников наибольшая конверсия  
(Построить путь клиента по всем источникам который в результате приводит к продажам)

Метрики, которые бы рассчитывались: CTR, CR, CAC, ROI

Также можно было бы составить маркетинговую воронку по каждому источнику

### Декомпозиция задачи

1. Сбор данных о пользователях со всех рекламных источников
2. Предобработка при необходимости
3. Построение пути действий пользователей, которые ведут к продаже (как воронка, но просто путь)
4. Расчет маркетинговых метрик
5. Построение воронок по источникам
6. Оценка наиболее значимых рекламных каналов

## Вклад rich-контента

Rich-контент - это маркетинговый инструмент, который позволяет добавить в описание на карточке товара мультимедийные элементы, такие как изображения и видео.

Для оценки влияния нам бы потребовалось подобрать пары максимально аналогичных пар товаров (для более точного исследования желательно больше пары), но с противоположным содержанием карточек товаров - одни будут наполнены rich-контентом, вторые будут довольно минималистичны. Важно взять именно те товары, у которых был сопоставимый объем продаж. 

После подбора необходимых товаров оценить как изменился объем продаж у товаров с rich-контентом и их доля в сравнении с конкурентами. Также рассчитывать конверсию с просмотра карточки товара в продажу.

### Декомпозиция задачи

1. Определение аналогичных товаров
2. Сбор данных о продажах за прошедший месяц и текущий
3. Предобработка при необходимости
4. Исследование с расчетом вышеуказанных показателей
5. Оценка результата
